# Instruction Print en Python

L'instruction `print` peut être utilisée de plusieurs manières différentes :

- `print "Bonjour le monde"`
- `print "Bonjour", <variable contenant la chaîne>`
- `print "Bonjour" + <variable contenant la chaîne>`
- `print "Bonjour %s" % <variable contenant la chaîne>`

**Exemple:**

```
print "Bonjour le monde"
```

**Sortie:**
```
Bonjour le monde
```

En Python, on peut utiliser des guillemets simples, doubles ou triples pour définir une chaîne de caractères. On utilise généralement :

- Des guillemets simples pour un seul caractère,
- Des guillemets doubles pour une ligne,
- Des guillemets triples pour un paragraphe ou plus

**Exemple:**

```
print 'Salut'
```

**Sortie:**

```
Salut
```

**Exemple:**

```
print """Hello, World !

I Love Python."""
```

**Sortie:**
```
Hello, World !

I Love Python.
```

Les chaînes peuvent être assignées à des variables, disons `chaine1` et `chaine2`, qui peuvent ensuite être appelées dans une instruction `print`.

**Exemple:**

```
chaine1 = 'World'
print 'Hello', chaine1

chaine2 = '!'
print 'Hello', chaine1, chaine2
```

**Sortie:**

```
Hello World
Hello World !
```


La concaténation de chaînes est l'« addition » de deux chaînes. Lors de la concaténation, il n'y aura pas d'espace entre les chaînes.

**Exemple:**

```
print 'Hello' + chaine1 + chaine2
```

**Sortie:**

```
HelloWorld!
```

**Utilisation de `%s` pour les variables:** `%s` est utilisé pour référer à une variable contenant une chaîne de caractères.

**Exemple:**

```
print "Hello %s" % chaine1
```

**Sortie:**

```
Hello World
```

**Utilisation avec d'autres types de données :**

- **`%s`** -> Chaîne de caractères (string)
- **`%d`** -> Entier (integer)
- **`%f`** -> Nombre décimal (float)
- **`%o`** -> Octal
- **`%x`** -> Hexadécimal
- **`%e`** -> Exponentiel

Ces spécificateurs peuvent être utilisés pour effectuer des conversions directement dans l'instruction **`print`**.

**Exemple:**

```
print "Nombre réel = %d" % 18
print "Décimal du nombre = %f" % 18
print "Équivalent octal du nombre = %o" % 18
print "Équivalent hexadécimal du nombre = %x" % 18
print "Équivalent exponentiel du nombre = %e" % 18
```

**Sortie:**

```
Nombre réel = 18
Décimal du nombre = 18.000000
Équivalent octal du nombre = 22
Équivalent hexadécimal du nombre = 12
Équivalent exponentiel du nombre = 1.800000e+01
```

Lorsque plusieurs variables sont référencées, des parenthèses sont utilisées.

**Exemple :**

```
print "Hello %s %s" % (chaine1, chaine2)
```

**Sortie:**

```
Hello World !
```

**Autres exemples d'utilisation de print :**

- Affichage du symbole `%` :

  **Exemple:**

  ```
  print "Je veux %%d être imprimé %s" % 'ici'
  ```

  **Sortie:**

  ```
  print "Je veux %%d être imprimé %s" % 'ici'
  ```

- Répétition d'une chaîne :

  **Exemple:**

  ```
  print '_A' * 10
  ```

  **Sortie:**

  ```
  _A_A_A_A_A_A_A_A_A_A
  ```

- Saut de ligne :

  **Exemple:**

  ```
  print "Jan\nFév\nMar\nAvr\nMai\nJuin\nJuil\nAoût"
  ```

  **Sortie:**

  ```
  Jan
  Fév
  Mar
  Avr
  Mai
  Juin
  Juil
  Août
  ```

- Affichage d'un caractère spécial tel que `\n` :

  **Exemple:**

  ```
  print "Je veux \\n être imprimé."
  ```

  **Sortie:**

  ```
  Je veux \n être imprimé.
  ```

- Utilisation des tabulations `(\t)` et des sauts de ligne :

  **Exemple:**

  ```
  print """
  Routine :
  \t- Manger
  \t- Dormir\n\t- Répéter
  """
  ```

  **Sortie:**

  ```
  Routine :
	- Manger
	- Dormir
	- Répéter
  ```
